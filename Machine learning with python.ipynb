# Using mean,median and mode

```python
#assuming indian earnings has normal distribution of 25,000 with mean deviation of 15,000 with 5000 data points
import numpy as np

record=np.random.normal(25000,15000,5000)
np.mean(record)
```

```python
import matplotlib.pyplot as pp
pp.hist(record,50)#hist means histogram and 50 stands here for continuous data to discrete data 
pp.show()
```

```python
#the mean deviates a lot when a unique large value is added to the data set but median dosent shifts a lot,for ex.
np.median(record) 
```

```python
record=np.append(record,[100000])
np.mean(record)
```

```python
np.median(record)
```

# Mode

```python
#shows ages from 18 to 85 in random of 50 people
age=np.random.randint(18,high=85,size=50)
age
```

```python
from scipy import stats
stats.mode(age)
#it simply means that most common age is 54 and it appears 3 times
```

# Probality density function and Probability mass function

# Ordinal: Combination of both above type.

# PDF deals with continuous data and PMF with discrete data.

```python
#probability density function
from scipy.stats import norm
x=np.arange(-3,3,0.01)
pp.plot(x,norm.pdf(x))
```

```python
#exponential variation
from scipy.stats import expon
y=np.arange(0,10,0.01)
pp.plot(y, expon.pdf(y))
```

```python
#binomial prob mass function
from scipy.stats import binom
d=np.arange(0,10,0.01)
r,u=10,0.5
pp.plot(d,binom.pmf(d,r,u))
```

```python
#poisson pmf
from scipy.stats import poisson
mu=500
x=np.arange(400,600,0.5)
pp.plot(x,poisson.pmf(x,mu))
```
